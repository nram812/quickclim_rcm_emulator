```python
import numpy as np
import matplotlib.pyplot as plt

# Time axis: years from 2000 to 2100
years = np.arange(2000, 2101)

# Baseline warming trend
baseline = 0.02 * (years - 2000)

# Initialize figure
fig, axs = plt.subplots(1, 3, figsize=(15, 5), sharey=True)

# 1. Internal Variability
for i in range(10):
    noise = np.random.normal(0, 0.1, size=years.shape)
    axs[0].plot(years, baseline + noise, color="steelblue", alpha=0.6)
axs[0].set_title("Internal Variability")
axs[0].set_xlabel("Year")
axs[0].set_ylabel("Temperature Anomaly [°C]")
axs[0].grid(True)

# 2. Model Uncertainty
for i in range(10):
    model_offset = np.random.normal(0, 0.2)
    noise = np.random.normal(0, 0.1, size=years.shape)
    axs[1].plot(years, baseline + model_offset + noise, color="seagreen", alpha=0.6)
axs[1].set_title("Model Uncertainty")
axs[1].set_xlabel("Year")
axs[1].grid(True)

# 3. Scenario Uncertainty
for i in range(3):  # Three scenarios
    scenario_slope = 0.015 + i * 0.01  # Different scenario slopes
    for j in range(3):  # Internal variability within each scenario
        noise = np.random.normal(0, 0.1, size=years.shape)
        axs[2].plot(years, scenario_slope * (years - 2000) + noise, color="orange", alpha=0.6)
axs[2].set_title("Scenario Uncertainty")
axs[2].set_xlabel("Year")
axs[2].grid(True)

plt.tight_layout()
plt.show()
```
